```python
from pathlib import Path

import pandas as pd
from plotnine import aes, geom_line, ggplot, labs, theme_minimal
```

```python
csv = sorted((Path.cwd() / ".." / "model_output").glob("*.csv"))[-1]
with csv.open("r") as f:
    data = pd.read_csv(f, names=("time", "S", "I", "R"))
data = data.melt(
    id_vars=["time"],
    value_vars=["S", "I", "R"],
    var_name="compartment",
    value_name="count",
)
data.head()
```

```python
(
    ggplot(data=data, mapping=aes(x="time", y="count", color="compartment"))
    + geom_line()
    + labs(title="SIR Model Simulation", x="Time", y="Number of Individuals")
    + theme_minimal()
)
```
